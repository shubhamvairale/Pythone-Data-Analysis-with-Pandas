 ## Multiple filtering

```python
film.loc[film.release_year>=2018,].head(3)
```

```python
film.loc[film.release_year>=2018,"title"].head()
```

```python
film[(
    film.release_year>=2017)|(
    film.country=="United States")|(
    film.country=="Canada")].head(3)
```

```python
film[film.country.isin([
    "United States","Canada"])]  
```
